```python
# checking for outliers

# Select only numerical columns for outlier detection
numerical_df = df.select_dtypes(include=np.number)

# Calculate IQR for numerical columns
Q1 = numerical_df.quantile(0.25)
Q3 = numerical_df.quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = numerical_df[(numerical_df < lower_bound) | (numerical_df > upper_bound)].stack().reset_index().rename(columns={'level_0': 'row_index', 'level_1': 'column_name', 0: 'outlier_value'})


print("\nOutliers identified:")
display(outliers)

# Visualize outliers using box plots for a few numerical columns
numerical_cols = numerical_df.columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols[:10]): # Plotting first 10 numerical columns for demonstration
    plt.subplot(2, 5, i + 1)
    sns.boxplot(x=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

## **Data Cleaning and Preparation**

*  We will be dropping columns that has >50% missing
values.

*   We will fill the missing values using with either the mean or median
*   We will standerdize, rename columns with strange characters (for clarity and readability purpose)




Data Cleaning Plan steps:
Renaming Columns (Standardize Naming) We’ll convert column names to:

Lowercase

Snake case (_ instead of spaces or symbols)

We will shorten names for clearity where necessary.

```python
# cleaning and standerdizing column names
def clean_column_names(df):
    df.columns = (
        df.columns.str.strip()                              # Remove leading/trailing spaces
                  .str.lower()                              # Convert to lowercase
                  .str.replace(r"[^\w\s]", "", regex=True)  # Remove special characters
                  .str.replace(r"\s+", "_", regex=True)     # Replace spaces with underscores
    )
    return df

# Apply cleaning
df = clean_column_names(df)

# Show updated column names
print("Cleaned column names:\n")
print(df.columns.to_list())
```

Handling Missing Values,
These steps include:


* Inspecting missing values
* checking and understanding which features are most incomplete
* Deciding what to drop or fill (based on threshold or relevance)


```python
# Show missing values count per column
missing_values = df.isnull().sum()
print("\nMissing values per column:")
print(missing_values[missing_values > 0].sort_values(ascending=False)) # Only show columns with missing values
```

```python
# percentage of missing value per column
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nPercentage of missing values per column:")
print(missing_percentage[missing_percentage > 0].sort_values(ascending=False)) # Only show columns with missing values
```

**Decide How to Handle Missing Values**

| Missing % | Action                                                          |
| --------- | --------------------------------------------------------------- |
| > 60%     | Drop column (too much missing data, unreliable)                 |
| 30–60%    | Drop or consider imputing (depending on importance)             |
| < 30%     | Impute (fill in with mean/median/mode or forward/backward fill) |


Why drop Missing > 60?

These columns are mostly empty and likely to introduce noise if we try to fill them arbitrarily, also can significantly impact the quality and reliability of your analysis


```python
# droping columns with more than 60% missing values, print dropped colums length
columns_to_drop = missing_percentage[missing_percentage > 60].index
df.drop(columns=columns_to_drop, inplace=True)

df_cleaned = df.drop(columns=cols_to_drop)

print(f"Dropped {len(cols_to_drop)} columns due to high missing val.")
print(f"New shape: {df_cleaned.shape}")
```

```python
# checking for the remaining missing values
remaining_missing_values = df.isnull().sum()
print("\nRemaining missing values per column:")
print(remaining_missing_values[remaining_missing_values > 0].sort_values(ascending=False))
```

**Handle the remaining missing values (imputation).**

This process will be divided into three categories depending on how crucial and sparse the missing values are:

 Category 1: These columns have only one missing value. It is rare, so we will fill them using the   mean or median.

Category 2 These are columns with moderate missing values (e.g., < 400).  We'll use mean to fill them, unless there's a cause for mode (categorical) or median (skewed).  All of these are continuous.

Category 3: These Columns have missing values of (1,800 or more), which is approximately 60% missing values. If they are critical (e.g., water/sanitation access), we will consider filling them with either the mean or median or eliminated them completely.


```python
# Category 1: filling the missing values using the mean option
df_cleaned['existence_expectancy_at_birth'].fillna(df_cleaned['existence_expectancy_at_birth'].mean(), inplace=True)
df_cleaned['existence_expectancy_index'].fillna(df_cleaned['existence_expectancy_index'].mean(), inplace=True)
```

```python
# Category 2: filling the missing values using the mean option

columns_to_fill = ['gross_income_per_capita',
    'income_index',
    'expected_years_of_education_galactic_years',
    'mean_years_of_education_galactic_years',
    'intergalactic_development_index_idi',
    'education_index',
    'intergalactic_development_index_idi_rank']

for col in columns_to_fill:
    df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)
```

```python
# Category 3: filling the missing values using the mean option

df_cleaned['population_using_at_least_basic_sanitation_services_'].fillna(df_cleaned['population_using_at_least_basic_sanitation_services_'].mean(), inplace=True)
df_cleaned['population_using_at_least_basic_drinkingwater_services_'].fillna(df_cleaned['population_using_at_least_basic_drinkingwater_services_'].mean(), inplace=True)
```

```python
# print/display any remaining missing values
remaining_missing_values = df_cleaned.isnull().sum()
print("\nRemaining missing values per column:")
print(remaining_missing_values.sum())
```

**Brief Summary on how the Missing Data were Handled**

We filled all missing values using mean.

```python
# Re-select Top 10 and Bottom 10 Galaxies by IDI to ensure they are DataFrames
top_10_galaxies = df_cleaned.nlargest(10, 'idi').copy()
bottom_10_galaxies = df_cleaned.nsmallest(10, 'idi').copy()

# Add income and well-being for these comparison
top_10_galaxies['Group'] = "Top_10"
bottom_10_galaxies['Group'] = "Bottom_10"

# Combine as one dataframe
comparison_df = pd.concat([top_10_galaxies, bottom_10_galaxies])
```

```python
# Chart 1: IDI and well-being comparison against top 10 and buttion 10 galaxies

fig, ax = plt.subplots(1, 2, figsize=(14, 6))
comparison_df.groupby('Group')[['idi', 'well_being']].mean().plot(
    kind='bar', ax=ax[0], color=['#1f7777', '#ff7666'])
ax[0].set_title('Top 10 vs Bottom 10: IDI & Well-being')
ax[0].set_ylabel('Value')
ax[0].set_xticklabels(['Bottom 10', 'Top 10'], rotation=0)

# Chart 2: Income Per Capita against top 10 and buttion 10 galaxies
comparison_df.groupby('Group')['income_per_cap'].mean().plot(
    kind='bar', ax=ax[1], color=['#4ca664'])
ax[1].set_title('Top 10 vs Bottom 10: Income per Capita')
ax[1].set_ylabel('Income per Capita')
ax[1].set_xticklabels(['Bottom 10', 'Top 10'], rotation=0)

plt.tight_layout()
plt.show()
```

**Observations**

1. **IDI & Well-being (Left Chart)**

   * **Top 10 galaxies** show significantly higher IDI (close to 1) and better well-being compared to bottom 10.
   * Well-being difference is noticeable but not as extreme as IDI.

2. **Income per Capita (Right Chart)**

   * Top 10 galaxies have **much higher income levels**, highlighting economic strength as a key driver of development.
   * Bottom 10 galaxies lag far behind in income, which aligns with their low IDI and well-being scores.
